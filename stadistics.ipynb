```python
import json
import matplotlib.pyplot as plt

with open("./Data/videos_metadata.json", "r", encoding="utf-8") as file:
    data = json.load(file)

# Extraer datos
subclip_durations = []

for video in data["videos"]:
    subclips = video["video_metadata"]["subclips"]
    subclip_durations.extend([clip["duration"] for clip in subclips.values()])

# Generar histograma
plt.figure(figsize=(10, 6))
plt.hist(subclip_durations, bins=20, edgecolor='black', alpha=0.7)
plt.xlabel("Duración de los subclips (segundos)")
plt.ylabel("Frecuencia")
plt.title("Histograma de Duraciones de Subclips")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar el gráfico
plt.show()
```

```python
# Filtrar subclips con duración entre 1 y 12 segundos
filtered_subclips = [duration for duration in subclip_durations if 1 <= duration <= 9]

# Contar la cantidad de subclips en ese rango
num_filtered_subclips = len(filtered_subclips)

num_filtered_subclips
```
